```python
# Code to start you off and manipulate the data. .T is transpose - swap columns and rows
import pandas as pd

df = pd.read_csv("cost-of-living.csv", index_col=0).T.reset_index()
df.head()
```

## Rename column

```python
# your code here
df2 = df.rename(columns={"index":"location"})
df2.head(5)
```

## Get city and country

```python
# split on comma, limit to 1 split, expand returns a dataframe
df2[["city","country"]] = df2.location.str.split(", ", n=1, expand=True)
df2[["location","city","country"]].head()
```

```python
# And - if needed - correct for the US including states and nowhere else doing it
df2.country = df2.country.apply(lambda x: x if not "," in x else x.split(", ")[-1])
df2.head()
```

## Figure out which country has the most cities

```python
# your code here
# df2.groupby("country").count().sort_values(by="location", ascending=False)
df2.country.value_counts()
```

```python
most_cities = df2.country.value_counts().index[0]
most_cities
```

## Create a subset of only that country

```python
# your code here
df3 = df2[df2.country==most_cities].copy()
df3.head()
```

## Sort by housing accommodation

```python
col = "Apartment (1 bedroom) in City Centre"
# your code here
df3[["country","city",col]].sort_values(by=col, ascending=False)
```
